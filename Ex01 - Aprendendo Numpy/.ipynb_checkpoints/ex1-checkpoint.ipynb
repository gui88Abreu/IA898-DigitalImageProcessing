### Tabuleiro de Xadrez

Para aprimorar as habilidades utilizando **numpy arrays**, foi proposto gerar a imagem de um tabuleiro de xadrez. A minha solução foi gerar um array de 1's e em seguida zerar as posições especificas da matriz para que ela represente o tabuleiro.

### Solução para uma Imagem 8x8

Abaixo está a solução simplista para uma imagem 8x8.

```python
import numpy as np

height, width = 8, 8

gry_scl_img = np.ones((height,width))

gry_scl_img[1:height:2, 0:width:2] = 0.0
gry_scl_img[0:height:2, 1:width:2] = 0.0
    
```

Para mostrar um esboço da imagem que será gerada foi utilizada a biblioteca **matplotlib**.

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.title('Constant image')
plt.imshow(gry_scl_img, cmap = 'gray', vmin = 0.0, vmax = 1.0)
```

### Solução para uma imagem NxM

Certamente, em sistemas reais, uma imagem terá uma resulução muito maior que **8x8**. Por isso, logo abaixo está a solução que contempla qualquer resolução **NxM**.

```python
height, width = 3000, 3000 #define altura e largura da imagem

gry_scl_img = np.ones((height, width)) #cria matriz de 1's

x_rct = height//8 #encontra a altura do retangulo das posicoes do tabuleiro
y_rct = width//8 #encontra a larguara do retangulo das posicoes do tabuleiro

#preenche de preto os retangulos necessarios
for j in range(0,8,2):
    for i in range(0,8,2):
        gry_scl_img[(i+1)*x_rct:(i+2)*x_rct, (j)*y_rct:(j+1)*y_rct] = 0
        gry_scl_img[(j)*x_rct:(j+1)*x_rct, (i+1)*y_rct:(i+2)*y_rct] = 0
```

Aqui é mostrado um esboço da imagem de maior resolução

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.title('Constant image')
plt.imshow(gry_scl_img, cmap = 'gray', vmin = 0, vmax = 1)
```
