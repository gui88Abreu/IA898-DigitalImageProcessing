# Importando Bibliotecas

```python
import numpy as np
from PIL import Image
img1 = Image.open('cameraman.tif')
width, height = img1.size
```

# Resolução Espacial

### 256x256

```python
img1
```

### 128x128 pixels

```python
img2 = img1.resize(size=(width//2, height//2))
img2
```

### 64x64 pixels

```python
img3 = img1.resize(size=(width//4, height//4))
img3
```

### 32x32 pixels

```python
img4 = img1.resize(size=(width//8, height//8))
img4
```
